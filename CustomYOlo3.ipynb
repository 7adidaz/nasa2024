# Imports

```python
import cvzone
from ultralytics import YOLO
import cv2
import torch
```

# Loading Model to CUDA / GPU
## YOLO will download the model for you if you do not have it, and you wrote the name correctly. 
### (Ex.. yolov9m.pt) and it will be auto downloaded online   

```python
# load cuda
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python


#load model with device
model = YOLO('yolov9d1.pt').to(device)
#V8
#model = YOLO("yolov8m-drone.pt").to(device)
model.info()
```

# Train the model
## You can configure this, and it will always save the best trained results in the runs folder

```python
results = model.train(data="drones.yaml", epochs=1, imgsz=320, batch=64)
```

# Image Test

```python
image = cv2.imread('yoto06497.png')
result = model.predict(image)

for detection in result:
    boxes = sorted(detection.boxes, key=lambda b: b.xywh[0][2] * b.xywh[0][3])

    if boxes:
        box = boxes[-1]
        if box.conf[0].item() < 0.7:
            continue

        x, y, w, h = box.xywh[0]
        rect = int(x - w / 2), int(y - h / 2), int(w), int(h)

        # Calculate the distance (you may need to adjust this calculation based on your specific use case)
        distance = -(x - image.shape[1] / 2)

        # Print the distance (replace this with your desired action)
        print(f'Distance: {distance}')

        # Draw a rectangle around the detected drone
        cvzone.cornerRect(image, rect, l=5, rt=3)
# save image to disk
cv2.imwrite('output.jpg', image)
```

```python
import os
model = YOLO("YOLOv9d1.pt").to(device)
# loop through the images in folder named images use foreach image in the folder do not use range
images = [image for image in os.listdir("output")]
for img in images:
    # if file is rar skip
    if img.endswith(".rar"):
        continue
        # read the image
    image = cv2.imread("output/"+img)
    # predict the image
    result = model.predict(image)
    # loop through the result
    for detection in result:
        # GET 1ST BOX IF IT NOT OUT OF RANGE
        if len(detection.boxes) == 0:
            cv2.imwrite("output_uav_unlabeled/"+img, image)
            continue
        with open("output_uav/"+img+".txt", "w") as f:
            for box in detection.boxes:
                
                    # get the x, y, w, h
                    x, y, w, h = box.xywh[0]
                    # get the confidence
                    conf = box.conf[0].item()
                    # get the class
                    cls = box.cls[0].item()
                    # get the label
                    label = model.names[int(cls)]
                    # print the label
                    print(label)
                    # draw the rectangle
                    cv2.rectangle(image, (int(x - w / 2), int(y - h / 2), int(w), int(h)), (0, 255, 0), 2)
                    # put the label
                    cv2.putText(image, label, (int(x - w / 2), int(y - h / 2 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                    cv2.imwrite("output_uav/"+img, image)
                    img_width, img_height = image.shape[1], image.shape[0]
                    x_norm = x / img_width
                    y_norm = y / img_height
                    w_norm = w / img_width
                    h_norm = h / img_height
                    f.write(f"1  {x_norm} {y_norm} {w_norm} {h_norm}\n")   


```
